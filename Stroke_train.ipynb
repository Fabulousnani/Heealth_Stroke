```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/Fabulousnani/Heealth_Stroke/main/healthcare-dataset-stroke-data.csv")
```

```python
def clean_data(data):
    # Replace missing values with the mean of each column in: 'bmi'
    data = data.fillna({'bmi': data['bmi'].mean()})
    # Round column 'bmi' (Number of decimals: 1)
    data = data.round({'bmi': 1})
    # Rename column 'bmi' to 'body_mass_index'
    data = data.rename(columns={'bmi': 'body_mass_index'})
    # Drop column: 'id'
    data = data.drop(columns=['id'])
    return data
    

data = clean_data(data.copy())
data.head()
```

```python
# Code generated by Data Wrangler for pandas DataFrame

def clean_data(data):
    # Change column type to int64 for column: 'age'
    data = data.astype({'age': 'int64'})
    # Change column type to int64 for column: 'hypertension'
    data = data.astype({'hypertension': 'int64'})
    # Change column type to int64 for column: 'heart_disease'
    data = data.astype({'heart_disease': 'int64'})
    # Change column type to int64 for column: 'heart_disease'
    data = data.astype({'stroke': 'int64'})
    # Change column type to int64 for column: 'gender_Male'
    data = data.astype({'gender_Male': 'int64'})
    # Change column type to int64 for column: 'ever_married_Yes'
    data = data.astype({'ever_married_Yes': 'int64'})
    # Change column type to int64 for column: 'work_type_Never_worked'
    data = data.astype({'work_type_Never_worked': 'int64'})
    # Change column type to int64 for column: 'work_type_children'
    data = data.astype({'work_type_children': 'int64'})
    # Change column type to int64 for column: 'Residence_type_Urban'
    data = data.astype({'Residence_type_Urban': 'int64'})
    # Change column type to int64 for column: 'smoking_status_formerly smoked'
    data = data.astype({'smoking_status_formerly smoked': 'int64'})
    # Change column type to int64 for column: 'smoking_status_never smoked'
    data = data.astype({'smoking_status_never smoked': 'int64'})
    # Change column type to int64 for column: 'smoking_status_smokes'
    data = data.astype({'smoking_status_smokes': 'int64'})
    # Change column type to int64 for column: 'work_type_Private'
    data = data.astype({'work_type_Private': 'int64'})
    # Change column type to int64 for column: 'work_type_children'
    data = data.astype({'work_type_children': 'int64'})
    # Change column type to int64 for column: 'work_type_Self-employed'
    data = data.astype({'work_type_Self-employed': 'int64'})
    # Change column type to int64 for column: 'gender_Other'
    data = data.astype({'gender_Other': 'int64'})
    data = data.rename(columns={'work_type_Self-employed': 'work_type_Self_employed'})
    data = data.rename(columns={'smoking_status_formerly_smoked': 'smoking_status_formerly_smoked'})
    data = data.rename(columns={'smoking_status_never_smoked': 'smoking_status_never_smoked'})                            
    return data

data = clean_data(data.copy())
data.head()
```

## X =features y= target

```python
X=data.drop(['stroke'],axis=1)
y=data.stroke
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=0) 
```

# Modelling

####  xtrain = this database  set contains the predictor for this train dataset
####  ytrain =this is the target variable within the training set
####  xtest = this data set contains the predictors for the test dataset
####  ytest =  this is the target variable within the test set

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(xtrain, ytrain)
```

```python
prediction = model.predict(xtest)
```

```python
from sklearn  import metrics
```

```python
import numpy as np
print ('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, prediction)))
```

```python
# feature Importance
feat_importances = pd.Series(model.feature_importances_, index=xtrain.columns)
feat_importances.nlargest(10).plot(kind='barh', figsize=(20,10))
plt.show()
```
